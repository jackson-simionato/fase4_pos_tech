# Tratamento outliers

A variável de Rendimento Anual possui problemas com outliers, que podem ser tratados a partir de uma técnica envolvendo média e desvio padrão das variáveis.

```python
def clean_outliers(df, column):
    mean = df[column].mean()
    std = df[column].std()

    lim_inf = mean - (2 * std)
    lim_sup = mean + (2 * std)

    index_outliers = []

    for i, row in df.iterrows():
        if row[column] < lim_inf or row[column] > lim_sup:
            index_outliers.append(i)

    return index_outliers
```

## Rendimento anual

```python
outliers_rendimento = clean_outliers(df_cadastrados_clean, 'Rendimento_anual')
```

```python
df_cadastrados_clean_outliers = df_cadastrados_clean.drop(outliers_rendimento)
df_cadastrados_clean_outliers.reset_index(drop=True, inplace=True)
```
